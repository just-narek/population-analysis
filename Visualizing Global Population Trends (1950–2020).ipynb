# Population Data Visualization Project


## Goal
The purpose of this project is to explore global population trends over time using data visualization techniques.  
By analyzing population growth patterns, we aim to identify key insights and communicate them effectively through clear and engaging visuals.

**Tools used**: Python, Pandas, NumPy Matplotlib

I got the data from Kaggle



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('population_total.csv')
```

```python
df
```

```python
# Dropping Null Values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Now drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# Creating a Pivot  for Cleaner Analysis
df = df.pivot(index='country', columns='year', values='population')
```

```python
df
```

# 1. Line Chart: Population Trend of a Single Country
### Problem:
How has the population of Armenia (or any country) changed over time?

```python
# Population of Armenia
Population_of_Armenia = df.loc['Armenia']
plt.plot(Population_of_Armenia.index, Population_of_Armenia.values, marker='o')
plt.grid(True)
plt.title('Armneia')
plt.xlabel('Year')
plt.ylabel('Population')

plt.savefig('Armenia.png')
plt.show()
```

```python
# Population of Russia
Population_Of_Russia = df.loc['Russia']
plt.plot(Population_Of_Russia.index, Population_Of_Russia.values, marker='o')
plt.grid(True)
plt.tight_layout()
plt.title('Russia')
plt.xlabel('Year')
plt.ylabel('Population')

plt.savefig('Russia.png')
plt.show()
```

# 4. Histogram: Population Buckets
### Problem:
How many countries had populations in different ranges (e.g., <1M, 1M–10M, etc.)?

```python
df.head()
```

```python
def pop_category(x):
    if x < 1000000:
        return  'less than 1m'
    elif 1000000 < x < 10000000:
        return "1m-10m"
    elif 10000000 < x < 50000000:
        return  "10m-50m"
    elif 50000000 < x < 100000000:
        return "50m-100m"
    elif x > 100000000:
        return "more than 100m"
```

```python
df['range'] = df[2020].apply(pop_category)
```

```python
df
```

```python
counts = df['range'].value_counts().sort_index()
```

```python
# Prepare data for matplotlib
labels = counts.index.tolist()
values = counts.values.tolist()

plt.figure(figsize=(8,5))
plt.bar(labels, values, color='teal')

plt.xlabel('Population Range')
plt.ylabel('Number of Countries')
plt.title('Number of Countries by Population Range (Using if-else)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('Range.png')  # ✅ Save before show
plt.show()
```

# 5. Cumulative World Population Over Time

Problem:
What is the total world population trend from 1950 to 2020


```python
df2 = pd.read_csv('population_total.csv')
```

```python
df2
```

```python
total = df2.groupby('year')['population'].sum()
```

```python
total_sum = total.sum()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Your existing code
total = df2.groupby('year')['population'].sum()
total_sum = total.sum()

# Basic line plot
plt.figure(figsize=(12, 6))
plt.plot(total.index, total.values, linewidth=2, color='blue', marker='o', markersize=4)
plt.title('World Population Trend Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Enhanced version with formatting
plt.figure(figsize=(14, 8))
plt.plot(total.index, total.values/1e9, linewidth=2.5, color='darkblue', marker='o', markersize=5)
plt.title('World Population Growth (1950-2020)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (Billions)', fontsize=14)
plt.grid(True, alpha=0.3)

# Add some styling
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().set_facecolor('#f8f9fa')

# Format y-axis to show billions
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:.1f}B'))

# Add annotations for key milestones
if len(total) > 0:
    max_pop = total.max()
    max_year = total.idxmax()
    plt.annotate(f'Peak: {max_pop/1e9:.1f}B ({max_year})', 
                xy=(max_year, max_pop/1e9), 
                xytext=(max_year-5, max_pop/1e9+0.3),
                arrowprops=dict(arrowstyle='->', color='red', alpha=0.7),
                fontsize=10, color='red')

plt.tight_layout()
plt.show()

# Alternative: Bar chart for specific years
plt.figure(figsize=(12, 6))
years_subset = total.index[::5]  # Every 5th year
values_subset = total.loc[years_subset]

plt.bar(years_subset, values_subset/1e9, color='steelblue', alpha=0.7, width=3)
plt.title('World Population by Decade', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (Billions)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for year, pop in zip(years_subset, values_subset):
    plt.text(year, pop/1e9 + 0.1, f'{pop/1e9:.1f}B', 
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Summary statistics
print(f"Total population across all years: {total_sum/1e9:.1f} billion")
print(f"Average annual population: {total.mean()/1e9:.1f} billion")
print(f"Population range: {total.min()/1e9:.1f}B to {total.max()/1e9:.1f}B")
print(f"Total growth: {((total.iloc[-1] - total.iloc[0])/total.iloc[0]*100):.1f}%")

# Growth rate analysis
growth_rates = total.pct_change() * 100
plt.figure(figsize=(12, 6))
plt.plot(growth_rates.index[1:], growth_rates.values[1:], 
         linewidth=2, color='green', marker='s', markersize=4)
plt.title('Annual Population Growth Rate', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Rate (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
